```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ratings=pd.read_csv("data/ratings.csv")
ratings
```

## **Ratings Database**

```python
df1=ratings.groupby("movie_id").mean()
df1=df1.sort_values(by='rating',ascending=False)
df1
```

### **Top 20 High Rated Movies**

```python
plt.figure(figsize=(25,10))
sns.barplot(x=df1.head(20).index,y=df1.head(20)["rating"])
```

```python
rate=ratings["movie_id"].value_counts()
rate.head(20).index
```

### **Top 20 Mostly Viewed Movies**

```python

plt.figure(figsize=(25,10))
sns.barplot(x=rate.head(20).index,y=rate.head(20).values)
```

## **Items Database**

```python
items=pd.read_csv("data/items.csv")
items.columns
```

```python
Genres=items[[ 'Action',
       'Adventure', 'Animation', "Children's", 'Comedy', 'Crime',
       'Documentary', 'Drama', 'Fantasy', 'Film-Noir', 'Horror', 'Musical',
       'Mystery', 'Romance', 'Sci-fi', 'Thriller', 'War', 'Western']]
l=Genres.sum()
l
```

```python
len([ 'Action',
       'Adventure', 'Animation', "Children's", 'Comedy', 'Crime',
       'Documentary', 'Drama', 'Fantasy', 'Film-Noir', 'Horror', 'Musical',
       'Mystery', 'Romance', 'Sci-fi', 'Thriller', 'War', 'Western'])
```

**There Are 18 Different Genres Movie In the Items Database.**

### **BarPlot Of Movies in Different Genres**

```python
plt.figure(figsize=(25,10))
sns.barplot(x=l.index,y=l.values)
```

```python
lan=items["original_language"].value_counts()
lan
```

### **BarPlot Of Movies in Different Languages**

```python
plt.figure(figsize=(25,10))
### Log Scaled version better for visualisation
sns.barplot(x=lan.index,y=np.log(lan.values)+1)
```

### **PieChart Of Movies in Different Languages**

```python
plt.figure(figsize=(25,10))
plt.pie(np.log(lan.values)+1,labels=lan.index)
```
